```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("loan.csv")
```

```python
data['income_cat']=pd.cut(data['Income'],bins=[0,30000,60000,90000,120000,np.inf],labels=[1,2,3,4,5])

```

```python
data
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)

for train_set,test_set in split.split(data,data["income_cat"]):
    train_set=data.loc[train_set].drop("income_cat",axis=1)
    test_set=data.loc[test_set].drop("income_cat",axis=1)
    
```

```python
loan = train_set.copy()

loan_labels = loan.drop("Loan_Approved", axis=1)  
loan_predictor = loan["Loan_Approved"]                      
```

```python
train_set
```

```python
num_cols = loan_labels.drop("Employment_Status", axis=1).columns.tolist()

cat_cols=["Employment_Status"]
```

```python
num_cols
```

```python
cat_cols
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline([
    ('impute',SimpleImputer (strategy='median')),
    ('scaling',StandardScaler())
])

cat_pipeline=Pipeline([
    ('encoding',OneHotEncoder())
])
```

```python
fullpipeline=ColumnTransformer([
    ('num',num_pipeline,num_cols),
    ('cat',cat_pipeline,cat_cols)
])
```

```python
loan_prepared = fullpipeline.fit_transform(loan_labels)
```

```python

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(loan_prepared, loan_predictor)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree=DecisionTreeClassifier()
tree.fit(loan_prepared, loan_predictor)
```

```python
# uses sigmoid function
from sklearn.linear_model import LogisticRegression
logic=LogisticRegression()
logic.fit(loan_prepared, loan_predictor)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(
    model,
    loan_prepared,
    loan_predictor,
    cv=5,
    scoring="accuracy"
)
print("Scores:", scores)
print("Mean score:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(
    tree,
    loan_prepared,
    loan_predictor,
    cv=5,
    scoring="accuracy"
)
print("Scores:", scores)
print("Mean score:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(
    logic,
    loan_prepared,
    loan_predictor,
    cv=5,
    scoring="accuracy"
)
print("Scores:", scores)
print("Mean score:", scores.mean())
```

```python
X_train = test_set.drop("Loan_Approved", axis=1)
y_train = test_set["Loan_Approved"]

X_train_transformed = fullpipeline.fit_transform(X_train)
tree.fit(X_train_transformed, y_train)
```

```python
X_test = test_set.drop("Loan_Approved", axis=1)
y_test = test_set["Loan_Approved"]

X_test_transformed = fullpipeline.transform(X_test)

y_pred = tree.predict(X_test_transformed)
```

```python
import joblib
import os
# Assuming your model and pipeline are already trained
joblib.dump(tree, "loan_model.pkl")
joblib.dump(fullpipeline, "loan_pipeline.pkl")

print("Model and pipeline saved!")
```
